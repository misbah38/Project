## Seaborn


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# upload the data to read using pandas library
mtcars=pd.read_csv("mtcars.csv")
mtcars
```

```python
mtcars.info()
```

```python
#shape is used to determine np. rows and columns respectively
mtcars.shape
```

```python
# now we use barplot to check our data 
res=sns.barplot(mtcars['cyl'], mtcars['mpg'])
plt.title("BarChart for Cylinder used per miles per gallon")
plt.show
```

```python
# this can also be written as this way
res=sns.barplot(x="cyl", y="mpg", data=mtcars)
plt.show
```

```python
# now we use barplot to check our data 
res=sns.barplot(mtcars['model'], mtcars['gear'])
plt.show
```

```python
# this to change the color
res=sns.barplot(x="cyl",y="mpg", data=mtcars, color="yellow")
plt.show
```

```python
#but what if we want to give different color? we use palette
res=sns.barplot(x="am",y="mpg", data=mtcars, palette="rocket")
plt.show
```

```python
res=sns.barplot(x="gear",y="hp", data=mtcars, palette="Set1")
plt.show
```

```python
#vertical countplot
sns.countplot(x='carb',data=mtcars,palette="Set2")
```

```python
#horizontal count plot
sns.countplot(y='gear',data=mtcars,palette="Set3")
```

```python
sns.countplot(x='gear',hue='cyl',data=mtcars,palette="Set1")
```

```python
#Histogram
sns.displot(mtcars.mpg,bins=10,color='g')
```

```python
#distribution plot
sns.distplot(mtcars.mpg,bins=10,color='r')
```

```python
#heatmap is used to visualize data in matrix
#The more light color means the positve relation the darker the color the negative the relation 
sns.heatmap(mtcars.corr(),cbar=True,linewidths=0.5)
```

```python
# iris data is stored in sns by default 
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].value_counts()
```

```python
# Scatter plot
sns.scatterplot(x='sepal_length',y='petal_length',data=iris)
```

```python
# now we use hue to characterize each flower specie
# size catogrization
sns.scatterplot(x='sepal_length',y='petal_length',data=iris,hue='species',size='species')
```

```python
#pair plot
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species',palette='Set1')
```

```python
#lmplot
sns.lmplot(x='petal_length',y='petal_width',data=iris)
```

```python
sns.lmplot(x='petal_length',y='petal_width',hue='species',data=iris,markers=['o','^','*'])
```

```python
#boxplot is also known as whisker plot which is used for qualitative data
sns.boxplot(x='species',y='sepal_width',data=iris)
```
